# Visualizing Decision Trees with dtreeviz

## Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
import dtreeviz
```

## Read Data

```python
from sklearn.datasets import load_diabetes

diabetes_data = load_diabetes()
X = pd.DataFrame(data = diabetes_data['data'], columns=diabetes_data['feature_names'])
y = diabetes_data['target']
```

```python
diabetes_data
```

```python
X.head(10)
```

## Training a Decision Tree

```python
dtree_reg = DecisionTreeRegressor(max_depth=3)
dtree_reg.fit(X, y)
```

## Visualizing the Decision Tree

```python
viz_model = dtreeviz.model(dtree_reg,
                           X_train=X, y_train=y,
                           feature_names=list(X.columns),
                           target_name='diabetes')
```

```python
viz_model.view()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

viz_model.rtree_leaf_distributions()
```
